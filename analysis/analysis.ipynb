## Load Cleaned Data & Libraries

```python
# Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Clean data
accomodation_occupancy = pd.read_csv('../data/clean_data/accomodation_occupancy.csv')
activities = pd.read_csv('../data/clean_data/activities.csv')
demographics = pd.read_csv('../data/clean_data/demographics.csv')
international_survey = pd.read_csv('../data/clean_data/international_survey.csv')
locations = pd.read_csv('../data/clean_data/location.csv')
regional_tourism = pd.read_csv('../data/clean_data/regional_tourism.csv')
transport = pd.read_csv('../data/clean_data/transport.csv')

#set plot sizes
sns.set_context('poster', font_scale=0.4)
plt.rcParams['figure.figsize'] = [5, 3]
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 10
```

## Analysis (business questions)
### Question 1: What kind of tourism activity generates the highest income for Scotland?

Unfortunately filtering `All` from each columns doesn't work so it'll be best to test specific demographics..

```python
demographics.query("social_grade != 'All'").head(10)
```

```python
# filter all columns which 'all' is in all columns...
demographics_filtered = demographics.query("age != 'All' | marital_status != 'All' | gender != 'All' | employment_status != 'All' | children != 'All' | access_to_car != 'All' | social_grade != 'All'").copy()
demographics_filtered
```

To fix this summary problem this specific dataset has, it is probably best to record a new column call `demopgraphic` to record when `All` doesnt appear in any of the columns from `breakdown_of_domestic_tourism`:`social_grade`

```python
# Melting dataframes!!!
melted_demo = demographics_filtered.copy()

melted_demo = melted_demo.melt(['year','units','value','breakdown_of_domestic_tourism'], var_name='category', value_name='demographic').query("demographic != 'All'").reset_index()
melted_demo.drop(['index'], axis=1, inplace=True)
melted_demo
```

```python
# Queried data for graphs
#  Expenditure
#  overall demographics (no category)
demo_q3_1 = (
    melted_demo.query("breakdown_of_domestic_tourism == 'Expenditure'")
    .groupby('demographic')
    .value
    .sum()
    .sort_values(ascending=False)
    .reset_index(name = 'total_expenditure')
    .copy()
)

# expenditure by age_group
demo_q3_2 = (
    melted_demo.query("breakdown_of_domestic_tourism == 'Expenditure' & category == 'age'")
    .groupby('demographic')
    .value
    .sum()
    .reset_index(name = 'total_expenditure')
    .copy()
)

# age_group overtime
demo_q3_3 = (
    melted_demo.query("breakdown_of_domestic_tourism == 'Expenditure' & category == 'age'")
    .groupby(['demographic', 'year'])
    .value
    .sum()
    .reset_index(name = 'total_expenditure_per_year')
    .copy()
)


# expenditure by social_grade
demo_q3_4 = (
    melted_demo.query("breakdown_of_domestic_tourism == 'Expenditure' & category == 'social_grade'")
    .groupby(['demographic'])
    .value
    .sum()
    .reset_index(name = 'total_expenditure')
    .copy()
)

# Visits
#totals
demo_q3_visits_1 = (
    melted_demo.query("breakdown_of_domestic_tourism == 'Visits'")
    .groupby('demographic')
    .value
    .sum()
    .sort_values(ascending=False)
    .reset_index(name = 'total_visits')
    .copy()
)
#age
demo_q3_visits_2 = (
    melted_demo.query("breakdown_of_domestic_tourism == 'Visits' & category == 'age'")
    .groupby(['demographic'])
    .value
    .sum()
    .reset_index(name = 'total_visits')
    .copy()
)
#grade
demo_q3_visits_3 = (
    melted_demo.query("breakdown_of_domestic_tourism == 'Visits' & category == 'social_grade'")
    .groupby(['demographic'])
    .value
    .sum()
    .reset_index(name = 'total_visits')
    .copy()
)
#over time
#grade
demo_q3_visits_4 = (
    melted_demo.query("breakdown_of_domestic_tourism == 'Visits' & category == 'age'")
    .groupby(['demographic', 'year'])
    .value
    .sum()
    .reset_index(name = 'total_visits_per_year')
    .copy()
)
```

### Question 4: What type of locations receive the most visits?
 - #### 4.1 What kinds of locations do people spend the most in?

```python
# filter data for graphs
# EXPENDITURE
q4_locations = (
    locations.query("location_type not in ['All'] & breakdown_of_domestic_tourism == 'Expenditure'")
    .groupby('location_type')
    .value
    .sum()
    .sort_values(ascending=False)
    .reset_index(name = 'total_expenditure')
    .copy()
)

q4_locations_exp_yearly = (
    locations.query("location_type not in ['All'] & breakdown_of_domestic_tourism == 'Expenditure'")
    .groupby(['location_type','year'])
    .value
    .sum()
    .sort_values(ascending=False)
    .reset_index(name = 'total_expenditure_per_year')
    .copy()
)

#VISITS
q4_locations_visits = (
    locations.query("location_type not in ['All'] & breakdown_of_domestic_tourism == 'Visits'")
    .groupby('location_type')
    .value
    .sum()
    .sort_values(ascending=False)
    .reset_index(name = 'total_visits')
    .copy()
)

q4_locations_visits_yearly = (
    locations.query("location_type not in ['All'] & breakdown_of_domestic_tourism == 'Visits'")
    .groupby(['location_type','year'])
    .value
    .sum()
    .sort_values(ascending=False)
    .reset_index(name = 'total_visits_per_year')
    .copy()
)
```

```python
sns.barplot(data = q4_locations, x='location_type', y='total_expenditure', palette='colorblind')
sns.despine()
plt.xlabel("Location Type", fontsize = 12)
plt.ylabel("Total Expenditure (m)" , fontsize = 12)
plt.title('Total Expenditure by Location Type' , fontsize = 16);
```

In terms of expenditure by tourists Cities is by far the highest

```python
sns.lineplot(data=q4_locations_exp_yearly, x='year', y='total_expenditure_per_year', hue='location_type', palette='colorblind')
sns.despine()
plt.legend(bbox_to_anchor=(1.05, 0.8), loc='upper left', borderaxespad=0)
plt.xlabel("Year", fontsize = 12)
plt.ylabel("Total Expenditure (m)" , fontsize = 12)
plt.title('Annual Expenditure by Location Types' , fontsize = 16);
```

It appears that the spending in all the location types is pretty consistent over the years, with the countryside slightly falling from 2017 to 2019 which could possible suggest there's less attractions possibly in the countryside, however due to the dataset we're not able to determine the issue correctly.

```python
sns.barplot(data = q4_locations_visits, x='location_type', y='total_visits', palette='colorblind')
sns.despine()
plt.xlabel("Location Type", fontsize = 12)
plt.ylabel("Total Visits (millions)" , fontsize = 12)
plt.title('Total Visits by Location Type' , fontsize = 16);
```

In terms of visits to different location types Cities are very popular, however towns & villages receive a large volume of visits aswell

```python
sns.lineplot(data=q4_locations_visits_yearly, x='year', y='total_visits_per_year', hue='location_type', palette='colorblind')
sns.despine()
plt.legend(bbox_to_anchor=(1.05, 0.8), loc='upper left', borderaxespad=0)
plt.xlabel("Year", fontsize = 12)
plt.ylabel("Total Visits (millions)" , fontsize = 12)
plt.title('Annual Visits by Location Types' , fontsize = 16);
```

Compared to the expenditure over time graph, it appears that there isn't that much of a gap between cities and town/villages in terms of annual visitation over the years, as apposed the hghe difference in terms of money spent in those locations.

```python
# Melting dataframes!!!
melted_accom = accomodation_occupancy.copy()

melted_accom = melted_accom.melt(['year','units','value','accommodation_type','occupancy_type'], var_name='category', value_name='accomodation').query("accomodation != 'All'").reset_index()
melted_accom.drop(['index'], axis=1, inplace=True)
melted_accom
```

```python
#data for graphs
q7_accom_type = (
    melted_accom.groupby('accommodation_type')
    .value
    .mean()
    .sort_values(ascending=False)
    .reset_index(name = 'avg_occupancy_percentage')
    .copy()
)

q7_accom_type_yearly = (
    melted_accom.groupby(['accommodation_type','year'])
    .value
    .mean()
    .sort_values(ascending=False)
    .reset_index(name = 'avg_occupancy_percentage_by_year')
    .copy()
)

q7_accom_location = (
    melted_accom.query("category == 'location'")
    .groupby('accomodation')
    .value
    .mean()
    .sort_values(ascending=False)
    .reset_index(name = 'avg_occupancy_percentage')
    .copy()
)

q7_accom_location_yearly = (
    melted_accom.query("category == 'location'")
    .groupby(['accomodation','year'])
    .value
    .mean()
    .sort_values(ascending=False)
    .reset_index(name = 'avg_occupancy_percentage_by_year')
    .copy()
)
```

```python
sns.barplot(data = q7_accom_type, x='avg_occupancy_percentage', y='accommodation_type', palette='colorblind')
sns.despine()
plt.xlabel("Occupancy rate (%)" , fontsize = 12)
plt.ylabel("Accommodation Type", fontsize = 12)
plt.title('Average Occupancy Rates for Accommodation Types' , fontsize = 16);
```

```python
sns.lineplot(data=q7_accom_type_yearly, x='year', y='avg_occupancy_percentage_by_year', hue='accommodation_type', palette='colorblind')
sns.despine()
plt.legend(bbox_to_anchor=(1.05, 0.8), loc='upper left', borderaxespad=0)
plt.xlabel("Year", fontsize = 12)
plt.ylabel("Occupancy rate (%)" , fontsize = 12)
plt.title('Yearly Average Occupancy Rates for Accommodation Types' , fontsize = 16);
```

```python
sns.barplot(data = q7_accom_location, x='avg_occupancy_percentage', y='accomodation', palette='colorblind')
sns.despine()
plt.xlabel("Occupancy rate (%)" , fontsize = 12)
plt.ylabel("Location", fontsize = 12)
plt.title('Average Occupancy Rates for Location Types' , fontsize = 16);
```

```python
sns.lineplot(data=q7_accom_location_yearly, x='year', y='avg_occupancy_percentage_by_year', hue='accomodation', palette='colorblind')
sns.despine()
plt.legend(bbox_to_anchor=(1.05, 0.8), loc='upper left', borderaxespad=0)
plt.xlabel("Year", fontsize = 12)
plt.ylabel("Occupancy rate (%)" , fontsize = 12)
plt.title('Yearly Average Occupancy Rates for Location Types' , fontsize = 16);
```

Large urban areas have maintained above a 55% occupancy rate on average from 2013-2019, other urban areas has aso steadily increased this may be sign that more people visits more built up areas than rural areas. ### Question 8: Is there any way to predict spending and visits for Visit Scotland?

The current data in its current form doesn't work well with modelling since the data is heavily summarised and seperated which leads to varying r-squared values (high or low) when performing linear regression.

Due to the data not being suited for this type of task, if `Visit Scotland` would like to be able to benefit from avanced analytics using preditive models, they would have to alter how their data is collected/saved. Perhaps collect data more regularly like monthly and standardise a format which allows all the tables to be joined easily.

To possible give an answer to this main question to predict spending for visit scotland. Analysis was carried out on a different dataset `International Passenger Survey Scotland` where some machine learning techniques where applied to potentially predicts the expenditure of the passengers from different countries.

Linear regression was initially performed on this different dataset, which didnt perform too well only correctly predicting expenditure 50% of the time. So as an extra to this an Unsupervied learning technique (K-Means Clustering) was applied to the data, to label the data based on their spending habits. The entire workflow using this data can be found [here...](q8_prediction.ipynb)

Adding the clustering technique definitively improved the overall performance of the model for all countries going from 50% to 66% accuracy, to improve on this score one country was selected which accounted from the most expenditure, visits and nights in the entire dataset which was `USA`. Performing some predictive analysis the model performance was 74% accurate before clustering and 82% after.

So it shows there multiple techniques that can be used to predict potential spending for visitors to scotland providing the data is collected/stored favourably for machine learning to be applied on the data.
